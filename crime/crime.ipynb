```python
sa = pd.read_csv('c.csv')
economy = pd.read_json('economy.json')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = sa['year'],
                         y = sa['crimes_against_person'],
                         mode = 'lines+markers', name = 'crimes_against_person',
                         line = dict(color = "Orange", width = 2)))
fig.add_trace(go.Scatter(x = sa['year'], y = sa['contact_related_crimes'],
                         mode = 'lines+markers', name = 'contact_related_crimes',
                         line = dict(color = "Green", width = 2)))
fig.add_trace(go.Scatter(x = sa['year'],
                         y = sa['other_serious_crimes'],
                         mode = 'lines+markers',
                         name = 'other_serious_crimes',
                         line = dict(color = "Red", width = 2)))
fig.update_layout(title = 'Worldwide Covid-19 Cases',
                  xaxis_tickfont_size = 14,
                  yaxis = dict(title = 'Number of Cases'))

fig.show()
```

```python
fig = go.Figure(data=go.Choropleth(
    locations = economy['iso_three'],
    z = economy['gdp_capita'],
    text = economy['country'],
    colorscale = 'Blues',
    autocolorscale=False,
    reversescale=True,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_tickprefix = '$',
    colorbar_title = 'GDP<br>Billions US$',
))

fig.update_layout(
    title_text='2014 Global GDP',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations = [dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='Source: <a href="https://www.cia.gov/library/publications/the-world-factbook/fields/2195.html">\
            CIA World Factbook</a>',
        showarrow = False
    )]
)

fig.show()
```
