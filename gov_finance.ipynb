```python
deficit_df = pd.read_csv("dandf.csv")
```

```python
def gov_dept_perc_gdp():
    """Central government debt, total (% of GDP) [GC.DOD.TOTL.GD.ZS]"""
    df = deficit_df[["year", "country", "gov_dept_perc_gdp"]]
    return df
```
